```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)      #Higher degree equals more accurate curve
x_poly = poly_reg.fit_transform(x)            #It transforms x into 2D array of x^0, x^1, x^2, x^3 etc...
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, linear_reg.predict(x), color = 'blue')
plt.title('Salary Predictor(Linear Reggression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg2.predict(x_poly), color = 'blue')
plt.title('Salary Predictor (Polynomial Regression)')
plt.ylabel('Salary')
plt.xlabel('Position Level')
plt.show()
```

```python
print(linear_reg.predict([[6.5]]))                #predict() takes 2D array as argument.
test_poly = poly_reg.fit_transform([[6.5]])       #before giving x for prediction we have to transform it into 2D array of matrix of feature.
print(lin_reg2.predict(test_poly))
```
